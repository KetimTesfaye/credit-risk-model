```python
# Cell 1: Overview of the Data
import pandas as pd
import numpy as np

# Load the core Xente transaction dataset
df = pd.read_csv("../data/raw/data.csv")

print("=== DATASET DIMENSIONS ===")
print(f"Total Rows (Records): {df.shape[0]}")
print(f"Total Columns (Features): {df.shape[1]}")

print("\n=== DATA TYPES AND RECORD AVAILABILITY ===")
print(df.info())

print("\n=== PREVIEW OF FIRST 5 ROWS ===")
print(df.head())
```

```python
# Cell 2: Summary Statistics
print("=== NUMERICAL FEATURES SUMMARY ===")
# Transpose (.T) for easier reading in dense data frames
display(df.describe().T)

print("\n=== CATEGORICAL FEATURES SUMMARY ===")
# Capture unique counts, top classes, and frequency modes
display(df.describe(include=['object', 'category']).T)
```

```python
# Cell 3: Distribution of Numerical Features
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Plot continuous numeric metrics (skipping zero-variance ID variables if needed)
for col in num_cols:
    if col in ['CountryCode']: 
        continue # Skip flat, uninformative attributes
        
    plt.figure(figsize=(9, 4))
    sns.histplot(df[col], kde=True, color='royalblue', bins=40)
    plt.title(f'Probability Density Distribution: {col}', fontsize=12, fontweight='bold')
    plt.xlabel(col)
    plt.ylabel('Frequency Density')
    plt.tight_layout()
    plt.show()
```

```python
# Cell 4: Distribution of Categorical Features
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

for col in cat_cols:
    # Handle columns with very high cardinality by only focusing on the top 10 groups
    top_categories = df[col].value_counts().head(10)
    
    plt.figure(figsize=(10, 4))
    sns.barplot(x=top_categories.index, y=top_categories.values, palette='plasma')
    plt.title(f'Volume Count Distribution: Top 10 Categories in {col}', fontsize=12, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Total Transactions')
    plt.tight_layout()
    plt.show()
```

```python
# Cell 5: Correlation Analysis
plt.figure(figsize=(8, 6))

# Generate the Pearson correlation matrix for numeric indicators
corr_matrix = df[num_cols].corr()

# Plot the matrix as a clear visual heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, vmin=-1, vmax=1)
plt.title('Numerical Feature Inter-Correlation Heatmap', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Cell 6: Identifying Missing Values
missing_counts = df.isnull().sum()
missing_percentages = (missing_counts / len(df)) * 100

missing_summary = pd.DataFrame({
    'Missing Values Count': missing_counts,
    'Percentage (%)': missing_percentages
}).sort_values(by='Missing Values Count', ascending=False)

print("=== COMPLETE MISSING DATA AUDIT ===")
display(missing_summary)
```

```python
# Cell 7: Outlier Detection
# Choose financial variables where outlier variance heavily drives risk profiles
target_financials = [col for col in num_cols if col not in ['CountryCode']]

for col in target_financials:
    plt.figure(figsize=(9, 3))
    sns.boxplot(x=df[col], color='lightcoral', flierprops={'marker': 'o', 'markersize': 4, 'alpha': 0.5})
    plt.title(f'Outlier Boundary Box Plot for: {col}', fontsize=12, fontweight='bold')
    plt.xlabel(f'{col} Unit Value')
    plt.tight_layout()
    plt.show()
```

 Executive Summary of EDA Insights (Credit Risk Implications)

Based on the exploratory data analysis conducted across the Xente dataset, the following 4 insights will directly dictate our downstream feature engineering and credit risk modeling strategies to maintain Basel II compliance:

 1. Zero-Variance Dimensionality Bloat (`CountryCode`)
Insight: The `CountryCode` feature contains a uniform value of `256` across all ~95,000 records, representing a completely localized transaction ecosystem within Uganda. 
Risk Action: Because this feature contains zero variance, it offers zero information gain for any machine learning algorithm. We will completely drop `CountryCode` from the production pipeline to save computing resources and avoid model bloat.

 2. Extreme Financial Skewness and High Outlier Exposure
Insight: Continuous metrics like `Amount` and `Value` exhibit extreme positive (right) skewness. The vast majority of records consist of low-value micro-transactions, but a small tail of massive transaction values sits far outside the standard Interquartile Range (IQR) whiskers.
Risk Action: In a Buy-Now-Pay-Later (BNPL) framework, these outliers cannot be discarded as bad data; they represent outsized credit loss exposure for Bati Bank. To stabilize linear models (like Logistic Regression) and prevent gradient descent distortion in tree models (like XGBoost), we must apply a Logarithmic Transformation ($log1p$) or robust scaling during feature engineering.

3. Clear Categorical Concentration
Insight: The distribution of categorical fields reveals that transaction volumes are heavily concentrated within a few dominant product categories and specific merchant providers, while a long tail of categories contains fewer than 1% of transactions.
Risk Action: Leaving high-cardinality features with sparse categories intact will cause model overfitting. We will group the low-frequency categories into a single `"Other"` flag and apply Weight of Evidence (WoE) encoding to convert these strings into stable, highly interpretable risk metrics.

4. High Multicollinearity Between Volume and Financial Totals
Insight: The correlation matrix highlights a near-perfect linear relationship ($r > 0.85$) between raw transaction absolute values and specific pricing metrics. 
Risk Action: Including both highly correlated variables would violate the non-multicollinearity assumption of Logistic Regression, artificially inflating coefficient variance. We will select the single most robust monetary volume metric to serve as our primary operational indicator.
